### Behind the scenes: QR Decomposition


**Problem:** $M$ is not invertible.


**Solution:** The matrix **M** can be factored into a product

\begin{equation*}
    M \; = \; QR
      \; = \;
  \left(
    \begin{array}{ccc}
        &            &   \\
        & orthogonal &   \\
        &   matrix   &   \\
        &            &   \\
    \end{array}
  \right)
  \left(
    \begin{array}{ccc}
        &            &   \\
        &            &   \\
        &    upper   &   \\
        & triangular &   \\
        &            &   \\
        &            &   \\
    \end{array}
  \right)
\end{equation*}

\begin{equation*}
    M \; = \; QR
      \; = \;
  \left(
    \begin{array}{ccc}
        &            &   \\
        & orthogonal &   \\
        &   matrix   &   \\
        &            &   \\
    \end{array}
  \right)
  \left(
    \begin{array}{ccccc}
       * & * & * & * & * \\
         & * & * & * & * \\
         &   & * & * & * \\
         &   &   & * & * \\
         &   &   &   & * \\
         &   &   &   &   \\
    \end{array}
  \right)
\end{equation*}

\begin{equation*}
    Q^T \; Q \; = \; I
\end{equation*}

## Example: Implementation of `polyfit()`


`````python
%%octave

function result = mypolyfit(x,y,deg)

    % Assume that x, y are already column vectors.
    % deg == Degree of polynomial to fit.
    
    n = length(x);
    M = ones(n,1);
    
    for i = 1:deg
        M = [ M  x.^i ];
    end
    
    result = M \ y;
end
`````

`````python
%%octave

x
y

a = mypolyfit(x, y, 2)
`````

### Covariance matrix

\begin{equation*}
  C =
  \left(
    \begin{array}{ccc}
      \sigma_a^2      & \text{cov(a,b)}  & \text{cov(a,c)} \\
      \text{cov(a,b)} & \sigma_b^2       & \text{cov(b,c)} \\
      \text{cov(a,c)} & \text{cov(b,c)}  & \sigma_c^2      \\
    \end{array}
  \right)
\end{equation*}

**Error in $a$: $\;\;\;\;\;\sigma_a \; = \; \sqrt{\sigma_a^2}$**

<p>&nbsp;</p>

In principle, one can get $C$ from:

\begin{equation*}
  C = (M^T \; M)^{-1} \; \sigma_y^2
  \;\;\;\;\;\;
  \rightarrow
  \;\;\;\;\;\;
  \text{where } \sigma_y = \text{random error in }y.
\end{equation*}

<p>&nbsp;</p>

**Problem:** $\sigma_y$ is not known

**Solution:** approximate $\sigma_y$

\begin{equation*}
   \sigma_y^2
      \; \approx
      \; \frac{ |\vec{y} - M\vec{a}|^2 }{n - m}
      \; =
      \; \frac{ \Sigma_i | \; y_i - (a + b x_i + c w_i) \; |^2 
             }{ (\text{number of rows}) - (\text{number of cols}) }
      \; =
      \; \frac{ \Sigma_i | \; \text{error(}y_i\text{)} \; |^2 
             }{ \text{degrees of freedom} }
\end{equation*}



`````python
%%octave

function [P C] = mypolyfit(x,y,deg)
    
    %
    % Find the best-fit parameters
    %
    M = ones(size(x));
    
    for i = 1:deg
        M = [ M  x.^i ];
    end
    
    P = M \ y;
    
    %
    % Find the errors
    %
    [nrows ncols] = size(M);
    
    if nrows > ncols
        sigma2 = dot(y - M*P, y - M*P) / (nrows - ncols);
        C = inv(M' * M) * sigma2;
    else
        C = 0;
        warning('Could not compute the covariance matrix.')
    end
end
`````

`````python
%%octave

x
y

[a C] = mypolyfit(x,y,2)

errors = sqrt(diag(C))
`````

`````python
%%octave

[a C] = mypolyfit(x,y,1)

errors = sqrt(diag(C))
`````
