## Linear Algebra

View the problem as a matrix equation:

\begin{equation*}
  \left(
    \begin{array}{cc}
      1 & x_1   \\
      1 & x_2   \\
      1 & x_3   \\
      ... & ... \\
      1 & x_n
    \end{array}
  \right)
  \left[
    \begin{array}{c}
      a_0   \\
      a_1   
    \end{array}
  \right]
  =
  \left[
    \begin{array}{c}
      y_1   \\
      y_2   \\
      y_3   \\
       ...  \\
      y_n
    \end{array}
  \right]
  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
  X \; \vec{a} \; = \; \vec{y}
  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
  \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
\end{equation*}

**Problem:** $X$ is not invertible.

### QR Decomposition



```python
%%octave

X = - [ 1 2 3 ;
        4 5 6 ;
        1 3 4 ;
        1 2 2 ;
        3 1 3 ]
```

```python
%%octave

[Q R] = qr(X)
```

```python
%%octave

Q*R
```

```python
%%octave

Q' * Q
```
